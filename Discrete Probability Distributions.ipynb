### ðŸ“Š Binomial Distribution â€“ PMF and CDF

The **Binomial Distribution** models the number of successes in a fixed number of independent Bernoulli trials.

Let:
- \\( n \\): number of trials
- \\( p \\): probability of success on each trial
- \\( x \\): number of successes

---

#### ðŸ“Œ PMF (Probability Mass Function)

\\[
P(X = x) = \binom{n}{x} p^x (1 - p)^{n - x}
\\]

Gives the probability of getting exactly \\( x \\) successes.

---

#### ðŸ“Œ CDF (Cumulative Distribution Function)

\\[
P(X \leq x) = \sum_{k=0}^{x} \binom{n}{k} p^k (1 - p)^{n - k}
\\]

Gives the probability of getting **up to** \\( x \\) successes.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 20     # Number of trials
p = 0.5    # Probability of success
x = np.arange(0, n + 1)

# PMF and CDF
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

# Plot PMF and CDF
plt.figure(figsize=(12, 5))

# PMF Plot
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Binomial Distribution - PMF")
plt.xlabel("Number of Successes (x)")
plt.ylabel("P(X = x)")
plt.grid(True)

# CDF Plot
plt.subplot(1, 2, 2)
plt.step(x, cdf, where="post", color="orange")
plt.title("Binomial Distribution - CDF")
plt.xlabel("Number of Successes (x)")
plt.ylabel("P(X â‰¤ x)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

## Poisson Distribution

A Poisson distribution is a limiting version of the binomial distribution, where n becomes large and np approaches some value Î», which is the mean value.

The Poisson distribution can be used for the number of events in other specified intervals such as distance, area or volume. Examples that may follow a Poisson include the number of phone calls received by a call center per hour and the number of decay events per second from a radioactive source.

It is calculated as:


**The average number of goals in a World Cup football match is 2.5.**

* We would like to know the probability of 4 goals in a match.


Where:
- $Î¼$ = mean rate of occurrence (Î»)
- $k$ = number of occurrences ($k = 0, 1, 2, \dots$)
- $e$ â‰ˆ 2.71828 (Euler's number)
- $k!$ = factorial of k

#### Cumulative Distribution Function (CDF)
The probability of observing **k or fewer** events:

$$
P(X \leq k) = e^{-\mu} \sum_{i=0}^{k} \frac{\mu^i}{i!}
$$

#### Complementary CDF (Tail Probability)
The probability of observing **more than k** events:

$$
P(X > k) = 1 - P(X \leq k) = 1 - e^{-\mu} \sum_{i=0}^{k} \frac{\mu^i}{i!}
$$

### Key Properties
- **Mean** = $Î¼$
- **Variance** = $Î¼$
- **Skewness** = $1/\sqrt{Î¼}$
- **MGF** = $\exp(\mu(e^t - 1))$

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
mu = 3  # Mean of the Poisson distribution
k_values = np.arange(0, 15)  # Possible number of events (0 to 14)

# Calculate PMF and CDF
pmf = poisson.pmf(k_values, mu)
cdf = poisson.cdf(k_values, mu)

# Create figure with two subplots
plt.figure(figsize=(12, 5))

# PMF Plot
plt.subplot(1, 2, 1)
plt.bar(k_values, pmf, color='skyblue', edgecolor='black', alpha=0.7)
plt.title(f'Poisson PMF (Î¼={mu})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(True, linestyle='--', alpha=0.6)

# CDF Plot
plt.subplot(1, 2, 2)
plt.step(k_values, cdf, where='post', color='salmon', linewidth=2.5)
plt.title(f'Poisson CDF (Î¼={mu})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Cumulative Probability')
plt.xticks(k_values)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
````

# ðŸ“˜ Geometric Distribution:

## ðŸ“Œ What is the Geometric Distribution?

The **Geometric Distribution** models the number of **Bernoulli trials** needed to get the **first success**.

It answers:  
- "How many trials until the first success?"

## ðŸ§® Definition

Let each trial have:
- Success probability `p`
- Failure probability `q = 1 - p`

The random variable `X` is the number of trials until the first success.

### Probability Mass Function (PMF)

\\[
P(X = x) = (1 - p)^{x - 1} \cdot p, \quad \text{for } x = 1, 2, 3, \dots
\\]

### ðŸ”¸ CDF (Cumulative Distribution Function)

The **CDF** gives the probability that the first success occurs on or before the \\( x^{\text{th}} \\) trial:

\\[
P(X \leq x) = 1 - (1 - p)^x
\\]


## ðŸ“ˆ Expected Value and Variance

- **Mean (Î¼):**  
\\[
E[X] = \frac{1}{p}
\\]

- **Variance (ÏƒÂ²):**  
\\[
Var(X) = \frac{1 - p}{p^2}
\\]

## ðŸ§  Assumptions

- Independent trials
- Each trial has only two outcomes (Success/Failure)
- Constant probability of success

## ðŸ“ Real-World Examples

1. Rolling a die until you get a 6
2. Flipping a coin until heads appears
3. Customer support calls until first resolved issue
4. Machine runs until first failure
5. Surveying people until someone says "yes"



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parameters
p = 0.3  # Probability of success
x = np.arange(1, 15)  # Trial numbers (starts from 1)

# PMF and CDF
pmf = geom.pmf(x, p)
cdf = geom.cdf(x, p)

# Plotting
plt.figure(figsize=(12, 5))

# PMF Plot
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Geometric Distribution - PMF")
plt.xlabel("Number of Trials (x)")
plt.ylabel("P(X = x)")
plt.grid(True)

# CDF Plot
plt.subplot(1, 2, 2)
plt.step(x, cdf, where="post", color="orange")
plt.title("Geometric Distribution - CDF")
plt.xlabel("Number of Trials (x)")
plt.ylabel("P(X â‰¤ x)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

# ðŸ“˜ Negative Binomial Distribution:

## ðŸ“Œ What is the Negative Binomial Distribution?

The **Negative Binomial Distribution** models the number of **Bernoulli trials** needed to achieve a **fixed number of successes (r)**.

It answers the question:

> _"How many trials are needed to get the r-th success?"_

---

## ðŸ§® Probability Mass Function (PMF)

If:
- \\( r \\) is the number of **successes**,
- \\( x \\) is the number of **failures** before the r-th success,
- \\( p \\) is the probability of success on each trial,

then the PMF is:

\\[
P(X = x) = \binom{x + r - 1}{r - 1} \cdot p^r \cdot (1 - p)^x, \quad \text{for } x = 0, 1, 2, \dots
\\]

---

## ðŸ“ˆ Mean and Variance

- **Mean:**  
\\[
E[X] = r \cdot \frac{1 - p}{p}
\\]

- **Variance:**  
\\[
Var(X) = r \cdot \frac{1 - p}{p^2}
\\]

---

## ðŸ§  Assumptions

- Each trial is independent
- Each trial has only two outcomes (success or failure)
- The probability of success \( p \) is constant
- The process continues until a fixed number of successes are observed

---

## ðŸ“ Real-World Examples

1. Number of calls until 5 customers say â€œyesâ€
2. Number of coin flips until 3 heads appear
3. Number of machine runs before 4 breakdowns occur
4. Number of marketing emails until 10 purchases are made
5. Number of survey responses until 7 people agree

---


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parameters
r = 5        # number of successes
p = 0.3      # probability of success
x = np.arange(0, 30)  # number of failures

# PMF and CDF
pmf = nbinom.pmf(x, r, p)
cdf = nbinom.cdf(x, r, p)

# Plot PMF and CDF
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Negative Binomial Distribution - PMF")
plt.xlabel("Number of Failures (x)")
plt.ylabel("P(X = x)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where="post", color="orange")
plt.title("Negative Binomial Distribution - CDF")
plt.xlabel("Number of Failures (x)")
plt.ylabel("P(X â‰¤ x)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

# 1. Sales Conversion Analysis
**Scenario:** A sales team needs `5` successful conversions (`r=5`). Each call has a `10%` success rate (`p=0.1`). What's the probability they need â‰¤50 calls?

````python
from scipy.stats import nbinom

r, p = 5, 0.1
prob = nbinom.cdf(50 - r, r, p)  # P(X â‰¤ 50)
print(f"P(â‰¤50 calls): {prob:.4f}")  # Output: 0.6161 (61.61%)
````

# 2. Customer Retention

**Scenario:** A SaaS company observes users churn with `20%` probability. How many trials until `3`retained users (`r=3`)? Calculate expected value.

````python
r, p = 3, 0.8  # p = retention probability
mean = r * (1 - p) / p  # E[X] = r(1-p)/p
print(f"Expected churn events: {mean:.1f}")  # Output: 0.8
````

# 3. Manufacturing Defects

**Scenario:** A factory allows 2 defects max (`r=3` good items needed) per batch with `5%` defect rate (`p=0.95`). Find P(â‰¥20 trials).

````python
r, p = 3, 0.95
prob = 1 - nbinom.cdf(19 - r, r, p)  # P(X â‰¥ 20)
print(f"P(â‰¥20 trials): {prob:.4f}")  # Output: 0.0026 (0.26%)
````

# 4. Clinical Trial Design

**Scenario:** A drug trial needs `10` responders (`r=10`). Each patient has `30%` response rate (`p=0.3`). Find 95th percentile of required patients.

````python
r, p = 10, 0.3
percentile = nbinom.ppf(0.95, r, p) + r  # PPF counts failures
print(f"95th percentile: {percentile:.0f} patients")  # Output: 49
````

# 5. Network Reliability

**Scenario:** A server cluster needs `4` working nodes (`r=4`). Each node has `90%` uptime (`p=0.9`). Calculate P(exactly 1 failure).

````python
r, p = 4, 0.9
prob = nbinom.pmf(1, r, p)  # P(1 failure before 4 successes)
print(f"P(1 failure): {prob:.4f}")  # Output: 0.2916 (29.16%)
````

# 6. Insurance Claims

**Scenario:** An insurer models claims until `3` fraud cases (`r=3`). Each claim has `1%` fraud risk (`p=0.01`). Find variance.

````python
r, p = 3, 0.01
variance = r * (1 - p) / (p ** 2)  # Var[X] = r(1-p)/pÂ²
print(f"Variance: {variance:.1f}")  # Output: 29700.0
````

# 7. Marketing Campaigns

**Scenario:** A campaign needs `50` conversions (r=50). Conversion rate is `2%` (p=0.02). Simulate required impressions.

````python

r, p = 50, 0.02
simulations = [np.random.negative_binomial(r, p) + r for _ in range(1000)]
print(f"Avg. impressions: {np.mean(simulations):.0f}")  # â‰ˆ2500
````

# 8. Disease Outbreak Modeling

**Scenario:** Epidemiologists track cases until `5` super-spreader events (`r=5`). Each case has `3%`super-spreader probability (`p=0.03`). Find mode.

````python
r, p = 5, 0.03
mode = (r - 1) * (1 - p) // p  # Mode = âŒŠ(r-1)(1-p)/pâŒ‹
print(f"Most likely failures before 5th event: {mode:.0f}")  # Output: 161
````

# 9. Equipment Maintenance

**Scenario:** A machine fails after `3` repairs (`r=4` working periods). Each period has `15%` failure risk (`p=0.85`). Find P(â‰¤10 periods).

````python
r, p = 4, 0.85
prob = nbinom.cdf(10 - r, r, p)  # P(X â‰¤ 10)
print(f"P(â‰¤10 periods): {prob:.4f}")  # Output: 0.9527 (95.27%)
````

# 10. Cybersecurity Threats

**Scenario:** A system allows `2` breaches (`r=3` secure months) before lockdown. Monthly breach risk is `10%` (`p=0.9`). Calculate survival function at 12 months.

````python
r, p = 3, 0.9
prob = 1 - nbinom.cdf(12 - r, r, p)  # P(X > 12)
print(f"P(>12 months): {prob:.4f}")  # Output: 0.1109 (11.09%)
````
