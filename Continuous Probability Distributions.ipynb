# üìò Uniform Distribution:

## üìå What is a Uniform Distribution?

The **Uniform Distribution** is a type of probability distribution in which **all outcomes are equally likely** within a given interval.

There are two types:

1. **Discrete Uniform Distribution** ‚Äì Finite number of equally likely outcomes (e.g., rolling a die).
2. **Continuous Uniform Distribution** ‚Äì Infinite number of outcomes in a continuous range (e.g., any value between 0 and 1).

---

## üìä Probability Density Function (PDF) - Continuous Case

For a continuous uniform distribution between \( a \) and \( b \):

\\[
f(x) =
\begin{cases}
\frac{1}{b - a} & \text{if } a \leq x \leq b \\
0 & \text{otherwise}
\end{cases}
\\]

---

## üî¢ Cumulative Distribution Function (CDF)

\\[
F(x) =
\begin{cases}
0 & \text{if } x < a \\
\frac{x - a}{b - a} & \text{if } a \leq x \leq b \\
1 & \text{if } x > b
\end{cases}
\\]

---

## üßÆ Mean and Variance

- **Mean**:
\\[
E[X] = \frac{a + b}{2}
\\]

- **Variance**:
\\[
Var(X) = \frac{(b - a)^2}{12}
\\]

---

## üìå Properties

- Shape: Rectangle (flat)
- All intervals of the same length within the range are equally probable
- Memoryless: ‚ùå No
- Support: \\( x \in [a, b] \\)

---

## üí° Examples of Uniform Distribution

- Random number generators (e.g., `rand(0,1)`)
- Selecting a random time between 2 PM and 4 PM
- Picking a point randomly on a line segment

---

## üêç Python Functions

| Function               | Purpose                    | Example                 |
|------------------------|-----------------------------|--------------------------|
| `uniform.pdf(x, a, b)` | PDF at value x              | `PDF at x = 5`          |
| `uniform.cdf(x, a, b)` | CDF at value x              | `P(X ‚â§ 7)`              |
| `uniform.ppf(q, a, b)` | Quantile (inverse CDF)      | `Value at 90th percentile` |
| `uniform.rvs(a, b, size)` | Random sampling         | `Generate random values`|

---

## üîç Visualization Example (Python Code)

Use the following Python code in a **code cell** to visualize the PDF and CDF:



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters
a = 2
b = 8
x = np.linspace(0, 10, 1000)

# PDF and CDF
pdf = uniform.pdf(x, loc=a, scale=b-a)
cdf = uniform.cdf(x, loc=a, scale=b-a)

# Plotting
plt.figure(figsize=(12, 5))

# PDF Plot
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.title("Uniform Distribution - PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True)

# CDF Plot
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='green')
plt.title("Uniform Distribution - CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# üìò Normal (Gaussian) Distribution: Full Guide

## üìå What is the Normal Distribution?

The **Normal Distribution**, also known as the **Gaussian Distribution**, is a continuous probability distribution that is **bell-shaped and symmetric** about the mean.

It is the most commonly used distribution in statistics and natural sciences due to the **Central Limit Theorem**, which states that the sum of many independent random variables tends toward a normal distribution.

---

## üßÆ Probability Density Function (PDF)

The formula for the **PDF** of a normal distribution with mean \( \mu \) and standard deviation \( \sigma \) is:

\\[
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}
\\]

---

## üî¢ Cumulative Distribution Function (CDF)

The **CDF** is the probability that a normally distributed variable \\( X \\) will be less than or equal to a given value \\( x \\):

\\[
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt
\\]

This integral does not have a closed form but is computed numerically.

---

## üìä Mean, Median, Mode, Variance

- **Mean** \\( \mu \\): Center of the distribution
- **Median**: Equal to the mean
- **Mode**: Equal to the mean
- **Variance** \\( \sigma^2 \\): Spread of the distribution
- **Standard Deviation** \\( \sigma \\): Square root of the variance

---

## üìà Properties

- Symmetric around the mean \\( \mu \\)
- Bell-shaped curve
- 68‚Äì95‚Äì99.7 Rule:
  - 68% of data within \\( \mu \pm 1\sigma \\)
  - 95% within \\( \mu \pm 2\sigma \\)
  - 99.7% within \\( \mu \pm 3\sigma \\)
- Continuous and unbounded
- Area under the curve is 1

---

## üí° Real-World Examples

- IQ scores
- Heights of people
- Measurement errors
- Test scores
- Blood pressure levels

---

---

## üìä Visualization Example (Python Code)






```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0       # Mean
sigma = 1    # Standard deviation
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# PDF and CDF
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Plot
plt.figure(figsize=(12, 5))

# PDF Plot
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label="PDF", color="blue")
plt.title("Normal Distribution - PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True)

# CDF Plot
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label="CDF", color="green")
plt.title("Normal Distribution - CDF")
plt.xlabel("x")
plt.ylabel("P(X ‚â§ x)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Normal Distribution Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0
sigma = 1

# Generate data
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, mu, sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'N(Œº={mu}, œÉ¬≤={sigma**2})')
plt.title('Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

# Probability of Events (Area Under Curve)

```python
# Calculate probabilities
prob_less_than_neg1 = norm.cdf(-1, mu, sigma)
prob_between_1_and_2 = norm.cdf(2, mu, sigma) - norm.cdf(1, mu, sigma)
prob_greater_than_1 = 1 - norm.cdf(1, mu, sigma)

print(f"P(X < -1): {prob_less_than_neg1:.4f}")
print(f"P(1 < X < 2): {prob_between_1_and_2:.4f}")
print(f"P(X > 1): {prob_greater_than_1:.4f}")

# Visualize P(1 < X < 2)
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='N(0,1)')
x_fill = np.linspace(1, 2, 100)
plt.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), color='orange', alpha=0.5, 
                 label=f'P(1<X<2) = {prob_between_1_and_2:.3f}')
plt.legend()
plt.title('Probability Between Two Values')
plt.grid(True)
plt.show()
```

# Percentiles and Quantiles

```python
# Find percentiles
percentiles = [0.05, 0.25, 0.5, 0.75, 0.95]
values = norm.ppf(percentiles, mu, sigma)

for p, v in zip(percentiles, values):
    print(f"{p*100:.0f}th percentile: {v:.2f}")

# Find probability for given value
value = 1.5
prob = norm.cdf(value, mu, sigma)
print(f"\nP(X < {value}) = {prob:.4f} ({prob*100:.1f}%)")
```

# Sampling from Normal Distribution

```python
# Generate random samples
sample_size = 1000
samples = np.random.normal(mu, sigma, sample_size)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, density=True, alpha=0.7, label='Sample Histogram')
plt.plot(x, y, 'r-', lw=2, label='Theoretical PDF')
plt.title('Sampling from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Sample statistics
sample_mean = np.mean(samples)
sample_std = np.std(samples)
print(f"Sample mean: {sample_mean:.4f} (theoretical: {mu})")
print(f"Sample std: {sample_std:.4f} (theoretical: {sigma})")
```

# Central Limit Theorem Demonstration

```python
# Demonstrate CLT with uniform distribution
sample_size = 10000
num_samples = 500
sample_means = []

for _ in range(num_samples):
    samples = np.random.uniform(0, 1, sample_size)
    sample_means.append(np.mean(samples))

# Plot distribution of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, label='Sample Means')

# Theoretical normal distribution (CLT)
mu = 0.5  # mean of uniform(0,1)
sigma = np.sqrt(1/12) / np.sqrt(sample_size)  # std of mean
x = np.linspace(0.45, 0.55, 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2, label='Theoretical Normal')

plt.title('Central Limit Theorem Demonstration')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

# Quality Control (Six Sigma)

```python
# Manufacturing process example
process_mean = 10.0
process_std = 0.2
spec_lower = 9.5
spec_upper = 10.5

# Calculate defect probability
prob_defect = norm.cdf(spec_lower, process_mean, process_std) + \
              (1 - norm.cdf(spec_upper, process_mean, process_std))
sigma_level = norm.ppf(1 - prob_defect/2)

print(f"Defect probability: {prob_defect*100:.4f}%")
print(f"Sigma level: {sigma_level:.2f}œÉ")

# Visualization
x = np.linspace(process_mean - 4*process_std, process_mean + 4*process_std, 1000)
y = norm.pdf(x, process_mean, process_std)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Process (Œº={process_mean}, œÉ={process_std})')
plt.axvline(spec_lower, color='red', linestyle='--', label='Specification Limits')
plt.axvline(spec_upper, color='red', linestyle='--')
plt.fill_between(x, y, where=(x < spec_lower) | (x > spec_upper), 
                 color='red', alpha=0.3, label='Defect Area')
plt.title('Quality Control: Process Capability')
plt.xlabel('Measurement')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

# Hypothesis Testing (Z-test)

```python
# Example: Test if sample comes from population with mean 100
population_mean = 100
population_std = 15
sample = np.random.normal(105, population_std, 30)  # actual mean is 105

sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: sample mean is significantly different")
else:
    print("Fail to reject null hypothesis")
```

# Confidence Intervals

```python
# Calculate 95% confidence interval for population mean
sample_size = 50
sample = np.random.normal(100, 15, sample_size)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # ddof=1 for sample std

z_critical = norm.ppf(0.975)  # 95% CI
margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Visualization
plt.figure(figsize=(10, 4))
plt.errorbar(0, sample_mean, yerr=margin_of_error, fmt='o', 
             capsize=5, label='Mean with 95% CI')
plt.axhline(100, color='red', linestyle='--', label='True population mean')
plt.xlim(-1, 1)
plt.yticks(np.arange(90, 111, 2))
plt.title('Confidence Interval for Population Mean')
plt.legend()
plt.grid(True)
plt.xticks([])
plt.show()
```

# Multivariate Normal Distribution

```python
from scipy.stats import multivariate_normal

# Parameters
mu = [0, 0]
cov = [[1, 0.5], [0.5, 1]]  # covariance matrix

# Create grid
x, y = np.mgrid[-3:3:.1, -3:3:.1]
pos = np.dstack((x, y))
rv = multivariate_normal(mu, cov)

# Plot density
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.contourf(x, y, rv.pdf(pos), levels=20, cmap='viridis')
plt.colorbar()
plt.title('2D Normal Distribution Density')

# Generate samples
samples = rv.rvs(size=1000)
plt.subplot(122)
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('Samples from 2D Normal Distribution')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.tight_layout()
plt.show()
```
